```python
#importando bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importando dados

df = pd.read_csv("student_habits_performance.csv")
```

```python
#visualizando dados

df
```

```python
#tipo de dados

#colunas numÃ©ricas

cols = ['study_hours_per_day','social_media_hours','netflix_hours','exam_score','mental_health_rating','exercise_frequency' ]

#plotar o mapa de calor

sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm')
plt.title('correlaÃ§Ã£o entre hÃ¡bitos de estudo e desempenho em exames')
plt.show
```

```python
#alunos que estudam mais tem melhor desempenho?
#o grÃ¡fico de dispersÃ£o mostra uma correlaÃ§Ã£o positiva entre horas de estudo por dia e pontuaÃ§Ã£o em exames, indicando que alunos que estudam mais tendem a ter um desempenho melhor nos exames.



#grafico de dispersÃ£o com linha de regressÃ£o.



sns.lmplot(data=df,x='study_hours_per_day', y='exam_score')
plt.title('mais estudos âž” notas mais altas?')
plt.show()
```

```python
#comparando mÃ©dias de notas entre alunos que estudam mais de 5 horas por dia e menos de 2 horas por dia.

filtro_estudo_alto = df[df['study_hours_per_day'] > 5]
filtro_estudo_baixo = df[df['study_hours_per_day'] < 2]

grupo_estudo_alto = filtro_estudo_alto['exam_score'].mean()
grupo_estudo_baixo = filtro_estudo_baixo['exam_score'].mean()

print ('mÃ©dia de notas dos alunos que estudam mais de 5 horas por dia:', grupo_estudo_alto)
print ('mÃ©dia de notas dos alunos que estudam menos de 2 horas por dia:', grupo_estudo_baixo)
# --- IGNORE Ã‰ SO PRA DEIXAR BONITO ---
```

```python
#tempo gasto em rede socias vs notas

sns.histplot(data=df,x='social_media_hours')
plt.title('distribuiÃ§Ã£o de horas em redes sociais')
plt.show()
```

```python
#avaliando notas mÃ©dias
#por diferentes intervalos (bins) de tempo gasto em redes sociais
#('0-2' horas, '2-4' horas, '4-6' horas, '6h+')

df['social_media_bins'] = pd.cut(df['social_media_hours'], bins=[0,2,4,6,float('inf')], labels=['0-2h','2-4','4-6','6h+'])

#plotar grÃ¡fico em caixa (boxplot)

sns.boxplot(data=df, x='social_media_bins', y='exam_score')
plt.title('notas mÃ©dias por tempo gasto em redes sociais')
plt.show()
```

```python
#frequÃªncia de exercÃ­cios fÃ­sicos vs notas em exames


#plotar grÃ¡fico em caixa (boxplot)

sns.boxplot(data=df, x='exercise_frequency_bins', y='exam_score')
plt.title('frequÃªncia de exercÃ­cios fÃ­sicos vs notas em exames')
plt.show()

#hÃ¡bitos saudÃ¡veis vs notas em exames

sns.boxplot(data=df, x='diet_quality', y='exam_score')
plt.title('qualidade da dieta vs notas em exames')
plt.show()
```

```python
#frequencia de exercicios fisicos
for col in ['exercise_frequency','mental_health_rating','diet_quality']:
    sns.boxplot(data=df, x=col, y='exam_score')
    plt.title(f'{col.replace("_"," ").title()} vs notas em exames')
    plt.show()
```
